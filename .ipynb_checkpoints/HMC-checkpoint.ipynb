```python
"""
    Create in 1 July 2018 at Cat & Cloud Coffee in Santa Cruz.
    Authur: Rui Meng.
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats, optimize 

class HMC():
    # Hamiltonian MCMC sampler
    
    def __init__(self, P_star=None, verbose=True, eps0 = 1e-5):
        self.verbose = verbose
        self.P_star = P_star
        self.auxi_pars = None
        self.eps0 = eps0
        
    def set_hyperpars(self, eps=1.0, L=10):
        self.eps = eps
        self.L = L
        
    def set_kwargs(self, **kwargs):
        if kwargs is not None:
            self.other_pars = kwargs
        else:
            self.other_pars = None
        
    def Comp_E(self, pars):
        if self.other_pars is None:
            return (-np.log(self.P_star(pars)))
        else:
            return (-np.log(self.P_star(pars, **self.other_pars)))
    
    def Comp_K(self, auxi_pars):
        return (np.matmul(auxi_pars.T, auxi_pars)/2.0)
    
    def Comp_E_grad(self, pars):
        return (optimize.approx_fprime(pars, self.Comp_E, self.eps0 * np.ones_like(pars)))
            
    def Comp_H(self, pars, auxi_pars):
        return (self.Comp_E(pars) + self.Comp_K(auxi_pars))
        
    def Leap_frog(self, pars, auxi_pars, eps, L):
        new_pars = pars
        new_auxi_pars = auxi_pars
        # print ("leap_frog start with initial values {}, {}".format(pars, auxi_pars))
        for l in range(L):
            new_auxi_pars -= self.eps/2.0*self.Comp_E_grad(new_pars)
            new_pars += self.eps*new_auxi_pars
            new_auxi_pars -= self.eps/2.0*self.Comp_E_grad(new_pars)
            # print (new_pars, self.Comp_E_grad(new_pars))

        return (new_pars, new_auxi_pars)
            
    def sample(self, pars, n_samples = 1):
        # Set the auxiliary variables at the first sampling time.
        self.pars = np.asarray(pars)
        # Initialize the random velocity
        self.auxi_pars = np.random.randn(*self.pars.shape)
        for i in range(n_samples):
            # print (self.pars, self.auxi_pars)
            new_pars, new_auxi_pars = self.Leap_frog(self.pars, self.auxi_pars, self.eps, self.L)
            # print (new_pars, new_auxi_pars)
            r = np.minimum(1, np.exp(self.Comp_H(self.pars, self.auxi_pars) - self.Comp_H(new_pars, new_auxi_pars)))
            # print (r)
            if np.random.uniform() < r:
                self.pars = new_pars
                self.auxi_pars = new_auxi_pars
        return (self.pars)
        
    
```

```python
import pandas as pd
import seaborn as sns
# Set the number of samplings 
n_samples = 1000
# Set all pars as a vector
pars = np.array([1.0])


def Cauchy(x):
    return (stats.cauchy.pdf(x))

def Normal(x, loc=0, scale=1):
    return (stats.norm.pdf(x,loc=loc,scale=scale))
    
    

HMC_cauchy = HMC(Normal)
HMC_cauchy.set_hyperpars()
HMC_cauchy.set_kwargs(loc = 10, scale = 10)

sampled_pars = np.zeros(n_samples)
for i in range(n_samples):
    pars = HMC_cauchy.sample(pars)
    sampled_pars[i] = pars
fig = plt.figure()
plt.plot(sampled_pars)
plt.title('traceplot for hmc')
fig = plt.figure()
plt.hist(sampled_pars, bins = int(len(sampled_pars)/5))
plt.title('histogram of standard cauchy distribution')
fig = plt.figure()
sns.distplot(sampled_pars, hist=True, kde=True)
plt.show()
    


```
